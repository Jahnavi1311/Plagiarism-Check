# Plagiarism Checker using TF-IDF & Cosine Similarity

#A Python program that compares text documents for similarity using **TF-IDF vectorization** and **cosine similarity**.  
#All `.txt` files inside the current directory are scanned, processed, and compared pairwise.

```python
# Import required libraries
import os                                          
from sklearn.feature_extraction.text import TfidfVectorizer   
from sklearn.metrics.pairwise import cosine_similarity        
```

```python
# Step 1: Collect all .txt files
student_files = [doc for doc in os.listdir() if doc.endswith('.txt')]

student_notes = [open(_file, encoding='utf-8').read() for _file in student_files]
```

```python
# Step 2: Convert text to TF-IDF vectors
def vectorize(Text):
    return TfidfVectorizer().fit_transform(Text).toarray()
```

```python
# Step 3: Compute cosine similarity
def similarity(doc1, doc2):
    return cosine_similarity([doc1, doc2])
```

```python
# Create vectors from text notes
vectors = vectorize(student_notes)

# Pair filenames with their vector representation
s_vectors = list(zip(student_files, vectors))

# A set to store plagiarism results without duplicates
plagiarism_results = set()
```

```python
# Step 4: Check plagiarism
def check_plagiarism():
    """
    Compares each document with every other document
    and calculates the similarity score.
    """
    global s_vectors

    # Iterate through all text files
    for student_a, text_vector_a in s_vectors:
        
        # Create a removable list copy
        new_vectors = s_vectors.copy()
        
        # Remove the current text from comparison
        current_index = new_vectors.index((student_a, text_vector_a))
        del new_vectors[current_index]
        
        # Compare with every other text
        for student_b, text_vector_b in new_vectors:
            
            # Calculate similarity score
            sim_score = similarity(text_vector_a, text_vector_b)[0][1]
            
            # Create sorted doc name pair for consistency
            student_pair = sorted((student_a, student_b))
            
            # Store similarity record
            score = (student_pair[0], student_pair[1], sim_score)
            plagiarism_results.add(score)
    
    return plagiarism_results
```

```python
# Step 5: Display plagiarism output
for data in check_plagiarism():
    print(data)    # Shows: (file1, file2, similarity_scor
```
